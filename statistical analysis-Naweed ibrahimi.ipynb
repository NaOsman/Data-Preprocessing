```python
import math
import statistics
import numpy as np
import scipy.stats
import pandas as pd
```

```python
x = [5,3,6,7,2,3,7,5,9,2,3]
x_with_nan = [8,6,7,25, math.nan,math.nan,math.nan, 4, 2.0]
```

```python
print(statistics.mean(x))
print(statistics.geometric_mean(x))
print(statistics.harmonic_mean(x))
print(statistics.median(x))
print(statistics.mode(x))
print(statistics.variance(x))
# print(statistics.stdev(x))


print(statistics.mean(x_with_nan))
print(statistics.geometric_mean(x_with_nan))
print(statistics.harmonic_mean(x_with_nan))
print(statistics.median(x_with_nan))
print(statistics.mode(x_with_nan))
print(statistics.variance(x_with_nan))
# print(statistics.stdev(x_with_nan))

```

```python
print(np.mean(x))
print(np.median(x))
print(np.var(x))
print(np.std(x))




print(np.nanmean(x_with_nan))
print(np.nanmedian(x_with_nan))
print(np.nanvar(x_with_nan))
print(np.nanstd(x_with_nan))
```

```python
print(scipy.stats.gmean(x))
print(scipy.stats.hmean(x))
print(scipy.stats.tmean(x))
print(scipy.stats.mode(x))
print(scipy.stats.tvar(x))
print(scipy.stats.tstd(x))

x_cleaned = [value for value in x_with_nan if not math.isnan(value)]


print(scipy.stats.gmean(x_cleaned))
print(scipy.stats.hmean(x_cleaned))
print(scipy.stats.tmean(x_cleaned))
print(scipy.stats.mode(x_cleaned))
print(scipy.stats.tvar(x_cleaned))
print(scipy.stats.tstd(x_cleaned))
```

```python

Covide_Died_case = np.array([[4, 5, 2],              
                              [8, 4, 2],              
                              [3, 5, 2],
                              [8, 27, 4],
                              [32, 5, 2]])

df = pd.DataFrame(Covide_Died_case, columns=['A', 'B', 'C'])

mean_col = df.mean()
median_col = df.median()
mode_col = df.mode().iloc[0]
variance_col = df.var()
covariance_col = df.cov()
percentiles_col = df.quantile([0.25, 0.5, 0.75])

mean_row = df.mean(axis=1)
median_row = df.median(axis=1)
mode_row = df.mode(axis=1).iloc[:, 0]
variance_row = df.var(axis=1)

print("Mean (Columns):\n", mean_col)
print("\nMedian (Columns):\n", median_col)
print("\nMode (Columns):\n", mode_col)
print("\nVariance (Columns):\n", variance_col)
print("\nCovariance (Columns):\n", covariance_col)
print("\nPercentiles (Columns):\n", percentiles_col)

print("\nMean (Rows):\n", mean_row)
print("\nMedian (Rows):\n", median_row)
print("\nMode (Rows):\n", mode_row)
print("\nVariance (Rows):\n", variance_row)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data
np.random.seed(0)
Covide_Died_case = np.random.randint(1, 50, size=(5, 3))
provience = ['Kabul', 'Kundoz', 'Herat', 'Mazar', 'Bamyan']
regions = ['Central', 'North', 'East']
df = pd.DataFrame(Covide_Died_case, index=provience, columns=regions)


```

```python
# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Box Plot of Covid Cases by Region')
plt.ylabel('Number of Cases')
plt.show()

```

```python
# Histogram
plt.figure(figsize=(10, 6))
df.plot.hist(alpha=0.7, bins=10)
plt.title('Histogram of Covid Cases')
plt.xlabel('Number of Cases')
plt.ylabel('Frequency')
plt.show()

```

```python
# Pie Chart
plt.figure(figsize=(8, 8))
df.sum().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Total Covid Cases by Region')
plt.ylabel('')
plt.show()

```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
df.plot.bar()
plt.title('Bar Chart of Covid Cases by Province and Region')
plt.ylabel('Number of Cases')
plt.show()

```

```python
# X-Y Plot (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Central', y='North', hue=df.index)
plt.title('X-Y Plot of Central vs North Cases')
plt.xlabel('Central Cases')
plt.ylabel('North Cases')
plt.show()

```

```python
# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Covid Cases')
plt.show()
```
